```python
import json 
import pandas as pd
import datetime as dt
from jsonschema import validate 
```

```python
file = open('users.json', mode = 'r')
Usercontents = file.read()
file.close()

Usercontents = Usercontents.replace('\n', ',')
Usercontents = "[" + Usercontents[:-1] + "]"
Usercontents
```

```python
file = open("UserFinal.json", "w", 1, "utf-8")
n = file.write(Usercontents)
file.close()
```

```python
with open('UserFinal.json', 'r') as f:
    data = json.loads(f.read())
    
dfUser = pd.json_normalize(data, max_level=1)
dfUser.head()
```

```python
dfUser = dfUser.rename(columns = {'active': 'Active', 'role': 'Role', '_id.$oid': '_Id', 
                       'createdDate.$date': 'CreatedDate', 'lastLogin.$date': 'LastLoginDate'})
```

# Code for Converting Brands into a CSV File

```python
file = open('brands.json', mode = 'r')
brandContents = file.read()
file.close()
```

```python
brandContents = brandContents.replace('\n', ',')
brandContents = "[" + brandContents[:-1] + "]"
brandContents = brandContents.replace("\\", "")
brandContents    
```

```python
file = open("brandsFinal.json", "w")
n = file.write(brandContents)
file.close()
```

```python
with open('brandsFinal.json', 'r') as f:
    data = json.loads(f.read())
    
dfBrands = pd.json_normalize(data, max_level=2)
dfBrands.head()
```

```python
dfBrands = dfBrands.rename(columns = {'_id.$oid': '_ID', 'cpg.$id.$oid': 'CPG_Id', 'cpg.$ref': 'CPG_Ref'})
dfBrands.dtypes
```

```python
dfBrands.head()
```

```python
dfBrands['barcode'] = dfBrands['barcode'].astype('int64')
dfBrands.dtypes    
```

```python
dfBrands.head()
```

```python
dfBrands.to_csv('BrandsFinal.csv')
```

# Code for Converting Receipts into a CSV

```python
file = open('receipts.json', mode = 'r')
receiptContent = file.read()
file.close()
```

```python
receiptContent = receiptContent.replace('\n', ',')
receiptContent = "[" + receiptContent[:-1] + "]"
receiptContent
```

```python
file = open("recriptFinal.json", "w")
n = file.write(receiptContent)
file.close()
```

```python
with open('recriptFinal.json', 'r') as f:
    data = json.loads(f.read())
    
dfReceipts = pd.json_normalize(data, max_level=2)
dfReceipts.head()
```

```python
dfReceipts.columns
```

```python
dfReceipts = dfReceipts.rename(columns = {'_id.$oid': '_Id', 'createDate.$date': 'CreatedDate', 
                                          'dateScanned.$date': 'DateScanned', 'finishedDate.$date': 'FinishedDate',
                                         'modifyDate.$date': 'ModifyDate', 'pointsAwardedDate.$date': 'PointsAwardedDate',
                                         'purchaseDate.$date':'PurchasedDate'})
dfReceipts.head()
```

```python
dfReceipts['CreatedDate'] = dfReceipts['CreatedDate'].fillna(0).astype('int64')
dfReceipts['CreatedDate'] = dfReceipts['CreatedDate'].apply(lambda d: dt.datetime.fromtimestamp(int(d) / 1000).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
dfReceipts.head()
```

```python
dfReceipts['DateScanned'] = dfReceipts['DateScanned'].fillna(0).astype('int64')
dfReceipts['DateScanned'] = dfReceipts['DateScanned'].apply(lambda d: dt.datetime.fromtimestamp(int(d) / 1000).strftime('%Y-%m-%d %H:%M:%S'))
dfReceipts.head()
```

```python
dfReceipts['FinishedDate'] = dfReceipts['FinishedDate'].fillna(0).astype('int64')
dfReceipts['FinishedDate'] = dfReceipts['FinishedDate'].apply(lambda d: dt.datetime.fromtimestamp(int(d) / 1000).strftime('%Y-%m-%d %H:%M:%S'))
dfReceipts.head()
```

```python
dfReceipts['ModifyDate'] = dfReceipts['ModifyDate'].fillna(0).astype('int64')
dfReceipts['ModifyDate'] = dfReceipts['ModifyDate'].apply(lambda d: dt.datetime.fromtimestamp(int(d) / 1000).strftime('%Y-%m-%d %H:%M:%S'))
dfReceipts['PointsAwardedDate'] = dfReceipts['PointsAwardedDate'].fillna(0).astype('int64')
dfReceipts['PointsAwardedDate'] = dfReceipts['PointsAwardedDate'].apply(lambda d: dt.datetime.fromtimestamp(int(d) / 1000).strftime('%Y-%m-%d %H:%M:%S'))
dfReceipts['PurchasedDate'] = dfReceipts['PurchasedDate'].fillna(0).astype('int64')
dfReceipts['PurchasedDate'] = dfReceipts['PurchasedDate'].apply(lambda d: dt.datetime.fromtimestamp(int(d) / 1000).strftime('%Y-%m-%d %H:%M:%S'))

dfReceipts.head()
```

```python
dfReceipts.to_csv('ReceiptsUnsure.csv')
```
